```python
#ライブラリの読み込み
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import matplotlib.pyplot as plt
import os.path
plt.style.use('ggplot')
plt.rcParams['font.family'] = 'IPAexGothic'
plt.rcParams['font.size'] = 16
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize'] = (15,10)
```

```python
#CSVファイルの読み込み，dfという名前のデータフレームにCSVファイルを読み込んでいます
df_all = pd.read_csv('../datasets/2010_2019_backup.csv')
df_creative = pd.read_csv('../datasets/creative_music.csv')

df_all = df_all.drop(['name', 'year', 'market', 'analysis_url', 'track_href', 'uri',  'type', 
                     'Unnamed: 0', 'Unnamed: 0.1', 'id','time_signature','mode','duration_ms','liveness'], axis=1)

df_creative = df_creative.drop(['analysis_url', 'track_href', 'uri',  'type', 'id','time_signature','mode','duration_ms','liveness'], axis=1)

df_all = df_all[~df_all.duplicated()]
df_all.duplicated()
```

```python
df_creative.head()
```

```python
df_all.head()
```

```python
#データのカラムのチェック
df = pd.concat([df_all, df_creative])
df = df[~df.duplicated(keep=False)]
df.info()
```

```python
sns.pairplot(df_creative, hue='genre')
plt.savefig('pariplot.png')
```

# 1. Histogram


```python
f, axes = plt.subplots(7, 2, figsize=(15, 30), sharex=False)
sns.despine(left=True)

for i, col in enumerate(df_creative.columns):
    sns.distplot(df_creative[col] ,kde=True, rug=False, hist=False, kde_kws={"shade": True}, color='c', ax=axes[int(i /2), i%2], label='creative music')
    sns.distplot(df_all[col] ,kde=True, rug=False, hist=False, kde_kws={"shade": True}, color='m',ax=axes[int(i/2), i%2], label='all')
plt.savefig('histogram.png')
```

```python
sns.scatterplot(x='energy', y='valence', data=df, hue='genre', palette='husl')
sns.scatterplot(x='acousticness', y='valence', data=df, hue='genre', palette='husl')
```

# 4. 物件の価格と建築年の分析
建築年ごとに物件の平均価格と標準偏差を表示しています．

```python
#ラインプロット（価格x建築年）
sns.lineplot(x="yr_built", y="price", data=df)
```

```python
#建築年ごとのデータ件数
df.groupby('yr_built').price.count().plot()
plt.ylabel('recorded number of built')
```

```python
df["log_price_bin"] = pd.cut(np.log(df["price"]),4) #物件価格を四つの価格帯にわける
var = 'log_price_bin'
sns.scatterplot(x='long', y='lat', data=df, hue=var, palette='husl')
```
